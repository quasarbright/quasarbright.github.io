```python
#setup
numDoors = 100
doors = [0 for x in range(numDoors)]
carind = randind(doors)
doors[carind] = 1
doors
```

```python
def getGuess(doors):
    return math.floor(random.random()*len(doors))
```

```python
#get which doors to reveal
def getOther(guess,doors):
    carind = doors.index(1)
    other = -1
    pool = list(range(len(doors)))#indices
    pool.pop(guess)
    if doors[guess] == 1:#car
        otherpind = randind(pool)#a random index in pool (index in an index array)
        other = pool.pop(otherpind)
    else:#goat
        pool = set(pool)
        pool.remove(carind)
        pool = list(pool)
        other = carind
    #     for pind in range(len(pool)):
    #         if doors[pool[pind]] == 1:
    #             other = pool.pop(pind)
    #             break
    return other,pool
```

```python
#end
def getResult(pool,guess,other,carind):
    print('goats are behind all of these doors: ',str(pool))
    print('that just leaves your guess, door number {guess}, and door number {other}. Do you switch?'.format(guess=guess,other=other))
    if other == carind:
        print('switch wins')
        return 'switch'
    elif guess == carind:
        print('stay wins')
        return 'stay'
```

```python
def simulate(guess,doors):
    other, pool = getOther(guess,doors)
    result = getResult(pool,guess,other,doors.index(1))
```
