```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def dydx(x,y):
    '''right side of the equation so dy/dx = dydx(x,y)'''
    return x+y
ic = (2,1)
dx = .001
X = []
Y = []
x = ic[0]
y = ic[1]
while x < 10 and y < 1000:
    X.append(x)
    Y.append(y)
    dy = dydx(x,y)*dx
#     print(x,y,dy)
    x += dx
    y += dy

plt.plot(X,Y)
```
