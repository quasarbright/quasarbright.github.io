```python
def linspace(a,b,n):
    '''returns an array of n numbers between and including a and b (evenly spaced)'''
    assert n > 1 and isinstance(n,int)
    assert b >= a
    dx = (b-a)/(n-1)
    return [a + x*dx for x in range(n)]
linspace(2,3,4)
```

```python
def nthderivative(f,n):
    ans = lambda x:f(x)#copy
    for i in range(n):
        ans = derivative(ans)
    return ans
X = linspace(-10,10,1000)
Y = [f(x) for x in X]
D1 = [nthderivative(f,1)(x) for x in X]
D2 = [nthderivative(f,2)(x) for x in X]
plt.plot(X,Y)
plt.plot(X,D1)
plt.plot(X,D2)
plt.grid(True)
```
